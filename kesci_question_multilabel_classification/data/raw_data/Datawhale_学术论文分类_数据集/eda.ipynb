```python
# pd.read_csv('train.csv')
with open('test.csv', 'r') as f:
    lines = f.readlines()
    papers = []
    cur_paper_lines = []
    # print(lines[0])
    for line in lines[1:]:
        if line.startswith('test_') and len(cur_paper_lines) > 0:
            papers.append(cur_paper_lines)
            cur_paper_lines = []
        cur_paper_lines.append(line)
    if len(cur_paper_lines) > 0:
        papers.append(cur_paper_lines)
```

```python
data = {'paperid':[], 'title':[], 'abstract':[], 
        # 'categories':[] 
        }
```

```python
import pandas as pd
def merge_lines(lines):
    lines = [line.strip() for line in lines]
    paper = "".join(lines)
    paper = paper.replace('\n', "").replace('\"', "")
    cols = paper.split('\t')
    if len(cols) != 3:
        print(lines)
    return cols
for p in papers:
    cols = merge_lines(p)
    data['paperid'].append(cols[0])
    data['title'].append(cols[1])
    data['abstract'].append(cols[2])
    # data['categories'].append(cols[3])
frame = pd.DataFrame(data)
frame
```

```python
frame.to_csv('test_data.csv', index=False)
```
